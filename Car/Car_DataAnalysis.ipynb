```python
def min_max(col):
    top = df[col].idxmax()  # returns the index of max value
    top_obs = pd.DataFrame(df.loc[top])  # returns the max value in the column with the data frame
    
    bottom = df[col].idxmin()  # returns the index of min value
    bottom_obs = pd.DataFrame(df.loc[bottom])  # returns the min value in the column with the data frame
    
    min_max_obs = pd.concat([top_obs, bottom_obs], axis = 1)
    return min_max_obs
```

```python
min_max('MPG_City')
```

```python
# Select all numerical variable
cars_num = df.select_dtypes(include =['float64', 'int64'])
cars_num
```

```python
cars_num.hist(bins=10)
```

```python
# correlation with the variable of interest
cars_corr = cars_num.corr()['MPG_City']
cars_corr
```

```python
# correlation plots using 'pairplot'
for i in range (0, len(cars_num.columns),5):
    sns.pairplot(cars_num, y_vars=['MPG_City'], x_vars= cars_num.columns[i:i+5])
```

```python
# plotting correlation in one plot (heatmap)
corr = cars_num.drop(['MPG_City','MPG_Highway'],axis=1.0).corr()
sns.heatmap (corr[(corr >= 0.5) | (corr <= 0.4)],
           cmap = 'viridis', vmax = 1.0, vmin = -1.0, linewidths = 0.1,
           annot = True, annot_kws = {'size' : 8}, square = True)
```
